```python
import re

def find_number_locations(lines):
    number_locations = []

    # Process each line
    for line_number, line in enumerate(lines, start=1):
        # Search for all sequences of digits and get their positions
        for match in re.finditer(r'\d+', line):
            number = int(match.group())
            x_position = match.start()
            y_position = line_number - 1  # Subtract 1 to make the index 0-based
            number_locations.append((number, x_position, y_position))

    return number_locations

# Example usage
values_block = [
    '............830..743.......59..955.......663..........................................367...........895....899...............826...220......',
    '.......284.....*............*.....$...+.....*...377..................*.......419.............488...*.......*...................*..-....939..',
    '....%.........976..679.461.7..........350..33.........$.380...$...151.897..........295..#......*....105.....418.............481........&....',
    '...992.....#......=...../........701................508...*..578........................259...331.................795..945........79........',
    '.........868........................*.............................17*..........348................441*852........*.....-...........@.....922',
    '....................*200............311..63................452.......323.#778.*....674....................680......696...372.....*..........',
    '.......266.......209......589.....=......*...365.........7.*...233.............755....*......644...272........697..*....*.....682..225......',
    '..836..........................949....607..........&....*..899...*....679.527.........331..........$....788../.....43....89.................',
    '........367.....328.&......%...............680...69..717.......60.......*.*............................*..........................728...264.',
    '........*.........*..119.253.......................................626....129...274............97...679.......752........*.......*....*.*...',
    '........360......471........../.........573-.702.............866...@...........*......772....../........259....*.........430....136.742.543.',
    '.........................../.852...............*...775...............643....455....../...........832....*.....41..535.......................',
    '.............-..........340.................103....*..........$...51*..................74................438......%...776...+.23*..663......',
    '.15.......806...............990.....................427....924............530.........=..............................*....968.....=....480..',
    '......#..............@914...*..............*........................%......*................................/.....406.................*.....',
    '....201.........79........592............70.894.....247..513+....367..$299.698......199................-.....22'
]

# Process the entire block and find number locations
locations = find_number_locations(values_block)

# Print the result
for number, x, y in locations:
    print(f"Number: {number}, X: {x}, Y: {y}")
```
