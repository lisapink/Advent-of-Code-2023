```python
import pandas as pd
```

```python
def prep_df(df):
    df[['dest_start','source_start', 'range']] = df['map:'].str.split(' ',expand=True)
    df['source_start'] = (df['source_start']).apply(lambda x: int(x))
    df['dest_start'] = (df['dest_start']).apply(lambda x: int(x))
    df['range'] = (df['range']).apply(lambda x: int(x))
    df['source_end'] = df['source_start'] + df['range']
    
    return df
```

```python
def perform_mapping(seed_list,df):
    output_list = []
    for seed_value in seed_list:
        for index, row in df.iterrows():
            if (seed_value >= row['source_start']) and (seed_value <= row['source_end']):
                diff = seed_value - row['source_start']
                output = diff + row['dest_start']
                output_list.append(output)
            else:
                continue
    if output is None:
        output = row['dest_start']
        output_list.append(output)
    return output_list
```

```python
def perform_mapping(seed_list, df):
    output_list = []
    
    for seed_value in seed_list:
        output = None  # Initialize output before the loop
        for index, row in df.iterrows():
            if (seed_value >= row['source_start']) and (seed_value <= row['source_end']):
                diff = seed_value - row['source_start']
                output = diff + row['dest_start']
                output_list.append(output)
            else:
                continue
        if output is None:
            # If no match is found in the inner loop, use a default value
            output = seed_value
            output_list.append(output)
    
    return output_list
```
