```python
import pandas as pd
```

```python
df = pd.read_csv('day8.csv')
df
```

```python
# Preprocessing the dataframe

df[['start', 'mapping']] = df['map'].str.split('=', expand=True)
df[['left','right']] = df['mapping'].str.split(',', expand=True)
df = df[['start', 'left','right']]
df['left'] = df['left'].str[1:]
df['left'] = df['left'].str[1:]
df['right'] = df['right'].str[1:-1]
df['start'] = df['start'].str[0:-1]
df
```

```python
directions = 'LRRLLRRLLRRRLRLRRLRLRLRRLRLRLRLLLRRRLRLRRLLRLRRRLRRRLRLRLRRRLLRRLLRLRLRRRLRLLRRRLRLLRLRRRLRLRRRLRRRLLRRRLLRRRLRRLRRRLLRRRLRRLRRLRRLLRLRRLRRLRRRLRLRLRLRLRLRRRLLRLRRLRLRRLLRRLLRLRRLRRRLRLRLRLRRRLRRLRRLRRLRRLRRLRLRRRLRRRLRRRLRLRLLRRLRRRLRRLRLRRRLRRLRRRLRRLLRRLLLRRRR'
directions_list = list(directions.strip(""))

new_starts_list = []
new_start = 416 #This is the position of AAA
counter = 1  

while True:
    for i in range(len(directions_list)):
        if directions_list[i] == 'L':
            new_key = df['left'][new_start]
            if new_key == 'ZZZ':
                print(f'success - Counter: {counter}')
                break
            matching_rows = df[df['start'].astype(str).str.strip("'") == new_key]
            if not matching_rows.empty:
                new_start_list = df.index[df['start'] == new_key].tolist()
                new_start = new_start_list[0]
                print(f'{new_key} - {new_start} - Counter: {counter}')
                
        else:
            new_key = df['right'][new_start]
            if new_key == 'ZZZ':
                print(f'success - Counter: {counter}')
                break
            matching_rows = df[df['start'].astype(str).str.strip("'") == new_key]
            if not matching_rows.empty:
                new_start_list = df.index[df['start'] == new_key].tolist()
                new_start = new_start_list[0]
                print(f'{new_key} - {new_start} - Counter: {counter}')

        counter += 1  
    else:
        continue
    break  
```
